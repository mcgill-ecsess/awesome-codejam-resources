```python
# OSMnx Demo

OSMnx is a python library allowing you to easily build street networks and analyze them.
```

```python
import osmnx as ox
import networkx as nx
import pandas as pd
```

## What can you do with OSMNX

* Graph from place
* Shape from place
* Graph from bbox
* Graph from point

### Useful links:
* https://geoffboeing.com/2016/11/osmnx-python-street-networks/ (overview)
* https://osmnx.readthedocs.io/en/stable/ (documentation)
* https://github.com/gboeing/osmnx-examples/ (more examples)

```python
ox.plot_graph(ox.graph_from_place("Verdun, Montreal, Canada", network_type='all'))
```

```python
ox.plot_shape(ox.gdf_from_place("Island of Montreal, Canada"))
```

```python
ox.plot_graph(ox.graph_from_bbox(45.52, 45.49, -73.55, -73.58))
```

```python
import osmnx as ox 
coord = (48.87378, 2.29504)
G = ox.graph_from_point(coord, distance=1000)
ox.plot_graph(G)
```

```python
# can get stats from graph we created
G = ox.graph_from_address("Arc de Triomphe, Paris")
```

```python
import json
stats = ox.basic_stats(G)
with open('results.txt', 'w') as json_file:
    json.dump(stats, json_file, indent=4, sort_keys=True)
```

```python
start_coord = (45.5049756, -73.5736905) # McGill University
end_coord = (45.5035380, -73.6176820) # Universite de Montreal
north = 45.5181450
south = 45.4854686
east = -73.5681800
west = -73.6279802

G = ox.graph_from_bbox(north, south, east, west, network_type='drive')

start_node = ox.get_nearest_node(G, start_coord)
end_node = ox.get_nearest_node(G, end_coord)

route = nx.shortest_path(G, start_node, end_node)
ox.plot_graph_route(G, route)
```
